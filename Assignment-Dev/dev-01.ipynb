## `Scenario 01`:

```python
import numpy as np


T = [[0.99, 0.2], [0.01, 0.8]]
T = np.array(T)

TotalPeople = 100_000
percentagePeopleIndustry = 0.3

peopleIndustry = percentagePeopleIndustry * TotalPeople
peopleAcademia = (1 - percentagePeopleIndustry) * TotalPeople

people = np.array([peopleIndustry, peopleAcademia])

initialYear = 2025
finalYear = 2125


people_history = [(initialYear, people)]
for year in range(initialYear + 1, finalYear + 1):
    people = T @ people
    people_history.append((year, people))
```

```python
X_axis = [year for year, _ in people_history]
Y_axis_industry = [people[0] for _, people in people_history]
Y_axis_academia = [people[1] for _, people in people_history]

Y_max = max(max(Y_axis_industry), max(Y_axis_academia)) * 1.1
Y_min = min(min(Y_axis_industry), min(Y_axis_academia)) * 0.8


X_ticks = np.linspace(initialYear, finalYear + 1, 21, dtype=int)
Y_ticks = np.linspace(Y_min, Y_max, 21)
```

```python
from matplotlib import pyplot as plt


plt.figure(figsize=(14, 6))
plt.plot(X_axis, Y_axis_industry, label='Industry')
plt.plot(X_axis, Y_axis_academia, label='Academia')

plt.xticks(X_ticks); plt.yticks(Y_ticks)
plt.ylim(Y_min, Y_max); plt.xlim(initialYear, finalYear)

plt.title('People in Industry vs Academia Over Time')
plt.xlabel('Year'); plt.ylabel('Number of People')

plt.legend(); plt.grid(alpha=0.3); plt.show()
```
